```python
def Decryption(path, output, password):
    with open(path, "rb") as EncryptedFile:
        pdf = PyPDF2.PdfReader(EncryptedFile)
        
        if pdf.is_encrypted:
            if pdf.decrypt(password):
                Writer = PyPDF2.PdfWriter()
                
                for page in range(len(pdf.pages)):
                    Writer.add_page(pdf.pages[page])
                    
                with open(output, "wb") as DecryptedFile:
                    Writer.write(DecryptedFile)
                    
                print("PDF has sucessfully Decrypted")
                
            else:
                print("Incorrect Password")
                
        else:
            print("PDF is not encrypted")
```

```python
Decryption('EncryptedFile.pdf', "EncryptedFile.pdf", "0123456789")
```
