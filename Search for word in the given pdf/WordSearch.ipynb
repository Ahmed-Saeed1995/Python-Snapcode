```python
import PyPDF2
```

```python
def WordSearch(OriginalFile, keyword):
    with open(OriginalFile, "rb") as file:
        pdf_reader = PyPDF2.PdfReader(file)
        
        for page in range(len(pdf_reader.pages)): # 0 - number of pages 
            PDFPages = pdf_reader.pages[ page ]
            text = PDFPages.extract_text()
            
            if keyword in text:
                print(F" the keyword {keyword} found in page num {page+1} \n {text}")
                return
        
        print(f"Keyword {keyword} not found in the file")
```

```python
WordSearch("English Simple Dictionary.pdf", "above board")
```

```python
def WordsSearch(OriginalFile, keyword):
    pdf_file = open(OriginalFile, "rb")
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    
    for page in range(len(pdf_reader.pages)): # 0 - number of pages
        PDFPage = pdf_reader.pages[ page ] # 0 - pages of file
        text = PDFPage.extract_text()
        
        if keyword in text:
            print(F" the keyword {keyword} found in page num {page+1}")
            next_page_number = page + 1
            
            while next_page_number < len(pdf_reader.pages):
                next_page = pdf_reader.pages[ next_page_number ] # page 2
                nex_text = next_page.extract_text()
                
                if keyword in nex_text:
                    print(F" the keyword {keyword} found in page num {next_page_number+1}")
                    next_page_number += 1
                    
                else:
                    break
```

```python
WordsSearch("Physics.pdf", "Physics")
```
